```python
# Import required libraries
from gensim.models import Word2Vec
from gensim.models.doc2vec import TaggedDocument
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the product data
data = pd.read_csv('https://raw.githubusercontent.com/geeky8/medrpha_recommendations/main/output.csv')

# Split category_name into separate words
data['categorystr'] = data['categorystr'].apply(lambda x: str(x).split())

# Define the function to tag and tokenize each product
def tag_products(data):
    tagged_products = []
    for index, row in data.iterrows():
        tagged_products.append(TaggedDocument(row['categorystr'] + [row['compnaystr']], [row['product_name']]))
    return tagged_products

# Tag and tokenize the products
tagged_products = tag_products(data)

# Define and train the Meta-Prod2Vec model
model = Word2Vec(min_count=1, window=3, sg=1, alpha=0.025, min_alpha=0.025, workers=4)
model.build_vocab(tagged_products)
model.train(tagged_products, total_examples=model.corpus_count, epochs=10, report_delay=1)

# Define the function to get similar products
def get_similar_products(product_name):
    similar_products = []
    product_vector = model.wv[product_name]
    for product in model.wv.vocab:
        if product != product_name:
            product_similarity = model.wv.similarity(product_name, product)
            if product_similarity > 0.5:
                similar_products.append((product, product_similarity))
    similar_products = sorted(similar_products, key=lambda x: x[1], reverse=True)
    return similar_products[:5]

# Get similar products for a given product
similar_products = get_similar_products('Avil 50')
print(similar_products)
```

```python
def voiceBasedNav(res):
  # import speech_recognition as sr
  # # import pyaudio

  # List of expected sentences
  expected_sentences = [
      "add to cart",
      "card mein add",
      "product nikaliye",
      "product add kare",
      "remove from cart",
      "go to cart",
      "checkout",
      "confirm checkout",
      "Avil 50",
      "Becosules cap",
      "Becadexamin cap",
      "Mahacef drop",
      "Ciplox 500",
      "Flexon syp",
      "I moon forte",
      "Shoot bolus",
      "I heal fast spray",
      "Safi 100 ml",
      "Kayam churan",
      "Pet safa churn",
      "Boroline cream",
      "Vicks vaporub",
      "Abhay tox",
      "Avaxim 80",
      "Bevac (adult)",
      "empty cart",
      "clear cart",

  ]

  def similarity(s1, s2):
      """
      Calculates the Levenshtein distance between two strings and returns a similarity score between 0 and 1.
      """
      m = len(s1)
      n = len(s2)

      if m == 0 or n == 0:
          return 0.0

      # Create a matrix to store the Levenshtein distances between substrings
      d = [[0 for j in range(n+1)] for i in range(m+1)]

      # Initialize the first row and column of the matrix
      for i in range(m+1):
          d[i][0] = i
      for j in range(n+1):
          d[0][j] = j

      # Fill in the rest of the matrix
      for j in range(1, n+1):
          for i in range(1, m+1):
              if s1[i-1] == s2[j-1]:
                  cost = 0
              else:
                  cost = 1
              d[i][j] = min(d[i-1][j] + 1, d[i][j-1] + 1, d[i-1][j-1] + cost)

      # Calculate the similarity score
      distance = d[m][n]
      similarity = 1 - (distance / max(m, n))

      return similarity

  # Find the closest match to the recognized sentence in the list of expected sentences
  best_match = max(expected_sentences, key=lambda x: similarity(x, res))

  # Print the closest match
  print("You said: " + res)
  print("Closest match: " + best_match)

  return best_match

voiceBasedNav("at 2 car")
```

```python
def getRecommendations(product):
  import pandas as pd
  from sklearn.metrics.pairwise import cosine_similarity
  from sklearn.feature_extraction.text import CountVectorizer

  # Load the dataset
  df = pd.read_csv('https://raw.githubusercontent.com/geeky8/medrpha_recommendations/main/output.csv')

  # create a count vectorizer object
  vectorizer = CountVectorizer()

  df['product_name'] = df['product_name'].str.strip()

  # compute the document-term matrix
  dtm = vectorizer.fit_transform(df['product_name'])

  # compute the cosine similarity matrix
  cosine_sim = cosine_similarity(dtm)

  # function to get top n similar products based on cosine similarity
  def get_similar_products(product_name, n=10):
      # get the index of the product_name in the dataset
      idx = df.index[df['product_name'] == product_name].tolist()[0]

      # get the cosine similarity scores of the product with all other products
      sim_scores = list(enumerate(cosine_sim[idx]))

      # sort the products based on cosine similarity scores
      sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

      # get the top n similar products
      sim_scores = sim_scores[1:n+1]

      # get the indices of the top n similar products
      product_indices = [i[0] for i in sim_scores]

      # return the top n similar products
      return df['product_name'].iloc[product_indices]

  return get_similar_products(product).tolist()


#   import pandas as pd
#   from sklearn.metrics.pairwise import cosine_similarity

#   # Load data
#   df = pd.read_csv('https://raw.githubusercontent.com/geeky8/medrpha_recommendations/main/output.csv')
#   df['product_name'] = df['product_name'].str.strip()

#   # Create user-item matrix
#   matrix = df.pivot_table(index='product_name', columns='order_id', values='quantity', fill_value=0)

#   # print(matrix)

#   # Calculate item-item similarity
#   item_sim = cosine_similarity(matrix.T)

#   # print(item_sim)

#   # Get k-most similar items to a given product
#   k = 5
#   product_name = product
#   product_index = matrix.loc[product_name]
#   similar_indices = item_sim[product_index].argsort()[::-1][:k+1]
#   # print(similar_indices)
#   similar_items = [(matrix.index[i], item_sim[product_index][i]) for i in similar_indices if (i != product_index).any() and (item_sim[product_index][i] > 0).any()]
#   similar_items
#   # Recommend other products
#   recommendations = []
#   for item, similarity in similar_items:
#       other_products = matrix.loc[item][matrix.loc[item] > 0].index
#       for product in other_products:
#           if product != product_name and product not in recommendations:
#               recommendations.append(product)

#   print('Recommendations for {}:'.format(product_name))
#   return recommendations[:10]


getRecommendations("B.N.C CREAM")
```
